```
df = pd.read_excel('Sample Walmart data.xlsx')
df
```

```
import re

def costum_tokenize(text):
  try:
    text = re.sub('\W+',' ', text)
    text = word_tokenize(text)
  except TypeError:
    text = 'NaN'
  
  return text
```

```
df['tokenized_SD'] = df['ShortDescription'].apply(costum_tokenize)
```

```
def word_count(word_list):
  return len(word_list)

df['short description wordcount'] = df['tokenized_SD'].apply(word_count)
```

```
df_impact = df[['Impact','short description wordcount']]
df_impact = df_impact.dropna()
```

```
wordcount_by_impact = df_impact.groupby(['Impact'])[['short description wordcount']].mean()

wordcount_by_impact
```
