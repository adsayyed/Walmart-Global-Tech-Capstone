```
def costum_tokenize(text):
  try:
    text = re.sub('\W+',' ', text)
    text = word_tokenize(text)
  except TypeError:
    text = 'NaN'
  
  return text

def freq_count(col):
  frq_dic = {}
  for lst in col:
    for i in lst:
      if i not in stop_words and i.isalpha():
        if i in frq_dic:
          frq_dic[i] += 1
        else:
          frq_dic[i] = 1
  return frq_dic

def freq_count2(col):
  frq_dic = {}
  for lst in col:
    for i in lst:
      if i not in stop_words:
        if i in frq_dic:
          frq_dic[i] += 1
        else:
          frq_dic[i] = 1
  return frq_dic
```

```
sub2['tokenized_SD'] = sub2['ShortDescription'].apply(costum_tokenize)
```

# New Section

```
frq = freq_count(sub2['tokenized_SD'])
df_frq = pd.DataFrame(columns = ['words', 'counts'])

df_frq['words'] = frq.keys()
df_frq['counts'] = frq.values()
```
